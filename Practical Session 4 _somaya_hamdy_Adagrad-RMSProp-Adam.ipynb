```python
def GDCostFunction(xc,yc,thetac):
    error=np.power(np.linalg.norm(((xc @ thetac.T)- yc),ord=2),2)
    mse=error/(2*len(xc))
    return mse
```

```python
def gradient(X,Y,Theta):
    theta_parameters=int(Theta.ravel().shape[1]) 
    temp = np.matrix(np.zeros(theta.shape))
    for j in range(theta_parameters):
        error = (X*Theta.T) - Y              # (summation(h(xi) - y ) 
        t= np.multiply(error,X[:,j])         # (summation(h(xi) - y ) *xi
        temp[:,j] = 1/len(x) * np.sum(t)
    grad=temp
    return grad
```

```python
def Adagrad(x,y,v,theta):
    
    if len(v)== 0:
        v.append(np.matrix([0,0]))
        return v

    v.append (v[-1] + np.power(gradient(x,y,theta),2))
    return v[-1]
```

```python
alpha=0.3
itera=1000
def Batch_Adagrad_GD(X,Y,Theta,lr,iteration):  
    temp = np.matrix(np.zeros(Theta.shape))
    cost = []                   
    theta_parameters=int(Theta.ravel().shape[1])     
    tt0=[]
    tt1=[]
    h=[]
    vt=[]
    eps=0.0000000000001
    for i in range(iteration):
        tt0.append(Theta[0,0])
        tt1.append(Theta[0,1])
        h.append(X*Theta.T)
     
        cost.append( GDCostFunction(X,Y,Theta) )

        vt.append(Adagrad(X,Y,v,Theta))
        vt_sqrt=np.sqrt(vt[-1])
        grad=gradient(X,Y,Theta)

        for j in range(theta_parameters):
        
            temp[:,j] = Theta[:,j] - (( lr* grad[:,j] ) /  ( vt_sqrt[:,j] + eps  ))
            
        Theta=temp
        
    return Theta , cost,tt0,tt1,h,vt
```

```python
def f(theta_0,theta_1,x):
    yhat = theta_0 + theta_1*x
    return yhat

def dx_theta_0(theta_0,theta_1,x,y):
    yhat = f(theta_0,theta_1,x)
    gradient = 2*(yhat - y)
    return gradient

def dx_theta_1(theta_0,theta_1,x,y):
    yhat = f(theta_0,theta_1,x)
    gradient = 2*x*(yhat - y)
    return gradient

```

```python
def Adagrad(x,y_2,alpha = 0.001,eps = 0.0000001,max_iteration = 100):
    theta_0 = np.random.uniform(0,1,1)
    theta_1 = np.random.uniform(0,1,1)
    
    G = [[0,0],
        [0,0]]
    iteration = 0
    while iteration < max_iteration:
        perm = np.random.permutation(len(x))
        xr = x[perm]
        yr = y_2[perm]
        x_x = xr[0]
        y_y = yr[0]
        
        
        
        g0 = dx_theta_0(theta_0,theta_1,x_x,y_y)
        g1 = dx_theta_1(theta_0,theta_1,x_x,y_y)
        
        G[0][0] = G[0][0] + g0 * g0 
        G[1][1] = G[1][1] + g1 * g1
        
        theta_0 = theta_0 - (alpha/np.sqrt(G[0][0]+eps)) * g0
        theta_1 = theta_1 - (alpha/np.sqrt(G[0][0]+eps)) * g1
        
        iteration += 1
        
        if iteration % 100 == 0 :
            print(f"iteration = {iteration}")
            print(f"theta_0 = {theta_0}")
            print(f"theta_1 = {theta_1}")
        return theta_0,theta_1
```

```python
def J(x,y_2,theta_0,theta_1):
    m=len(x)
    cost = (1/(2*m) *((theta_0 + (theta_1 *x )) - y_2)**2).sum()
    return cost


def dx_theta_0(theta_0,theta_1,x,y_2):
    yhat = theta_0 + theta_1*x
    gradient = 2*(yhat - y_2)
    return gradient

def dx_theta_1(theta_0,theta_1,x,y_2):
    yhat = theta_0 + theta_1*x
    gradient = 2*x*(yhat - y_2)
    return gradient
```

```python
def Adagrad(x,y_2,alpha = 0.0001,eps = 0.00000001,iterations = 10000):
    
    cost_list = []
    
    theta_0_list = []
    theta_1_list = []
    
    grad_0_list = []
    grad_1_list = []
    
    hypothesis_list=[]
    
    #theta intilization
    theta_0 = 0
    theta_1 = 0
    
    prev_v_theta_0 = 0
    prev_v_theta_1 = 0
    
    m = len(y_2)
    for i in range (iterations):
        
        #hypothesis
        h = theta_0 + (theta_1 * x)
        hypothesis_list.append(h)
        
        #get cost
        cost = J(x, y_2, theta_0 , theta_1)
        cost_list.append(cost)
        
        # caluulate gradiant
        
#         grad_0 = (( h - y ).sum())/m
#         grad_1= ((( h - y )* x) .sum())/m
        g0 = dx_theta_0(theta_0,theta_1,x,y_2)
        g1 = dx_theta_1(theta_0,theta_1,x,y_2)
        
        grad_0_list.append(g0)
        grad_1_list.append(g1)
        
        v_theta_0 = prev_v_theta_0  + g0 * g0
        v_theta_1 = prev_v_theta_0  + g1 * g1
        
        prev_v_theta_0 = v_theta_0
        prev_v_theta_1 = v_theta_1
        
        #update thetas
        theta_0 = theta_0 - (alpha/(np.sqrt(prev_v_theta_0)+eps )) * g0
        theta_1 = theta_1 - (alpha/(np.sqrt(prev_v_theta_1) + eps))  * g1
        
        theta_0_list.append(theta_0)
        theta_1_list.append(theta_1)
        
        
        #if ther's no change in the cost so break
        if i > 1:
            if abs(cost_list[i-1]- cost_list[i])<0.0001:
                break

            #if ther's no change in gradiant so, break
            if abs(np.linalg.norm([theta_0_list[i],theta_1_list[i]])-np.linalg.norm([theta_0_list[i-1],theta_1_list[i-1]]))< 0.0001:
                break

            #if no chane in thetas so break
            if abs(np.linalg.norm([grad_0_list[i],grad_1_list[i]])-np.linalg.norm([grad_0_list[i-1],grad_1_list[i-1]])) < 0.0001:
                break
    
    return theta_0 , theta_1, cost_list, theta_0_list, theta_1_list, hypothesis_list
```

## RMSProp

```python
def RMSPROP(x,y_2,alpha = 0.01,eps = 0.00000001,iterations = 10000):
    beta = .8
    cost_list = []
    
    theta_0_list = []
    theta_1_list = []
    
    grad_0_list = []
    grad_1_list = []
    
    hypothesis_list=[]
    
    #theta intilization
    theta_0 = 0
    theta_1 = 0
    
    prev_v_theta_0 = 0
    prev_v_theta_1 = 0
    
    m = len(y_2)
    for i in range (iterations):
        
        #hypothesis
        h = theta_0 + (theta_1 * x)
        hypothesis_list.append(h)
        
        #get cost
        cost = J(x, y_2, theta_0 , theta_1)
        cost_list.append(cost)
        
        # caluulate gradiant
        
#         grad_0 = (( h - y ).sum())/m
#         grad_1= ((( h - y )* x) .sum())/m
        g0 = dx_theta_0(theta_0,theta_1,x,y_2)
        g1 = dx_theta_1(theta_0,theta_1,x,y_2)
        
        grad_0_list.append(g0)
        grad_1_list.append(g1)
        
        v_theta_0 = beta*prev_v_theta_0  + (1-beta)*(g0 * g0)
        v_theta_1 = beta*prev_v_theta_1  + (1-beta)*(g1 * g1)
        
        prev_v_theta_0 = v_theta_0
        prev_v_theta_1 = v_theta_1
        
        #update thetas
        theta_0 = theta_0 - (alpha/(np.sqrt(prev_v_theta_0)+eps )) * g0
        theta_1 = theta_1 - (alpha/(np.sqrt(prev_v_theta_1) + eps))  * g1
        
        theta_0_list.append(theta_0)
        theta_1_list.append(theta_1)
        
        
        #if ther's no change in the cost so break
        if i > 1:
            if abs(cost_list[i-1]- cost_list[i])<0.0001:
                break

            #if ther's no change in gradiant so, break
            if abs(np.linalg.norm([theta_0_list[i],theta_1_list[i]])-np.linalg.norm([theta_0_list[i-1],theta_1_list[i-1]]))< 0.0001:
                break

            #if no chane in thetas so break
            if abs(np.linalg.norm([grad_0_list[i],grad_1_list[i]])-np.linalg.norm([grad_0_list[i-1],grad_1_list[i-1]])) < 0.0001:
                break
    
    return theta_0 , theta_1, cost_list, theta_0_list, theta_1_list, hypothesis_list
```

```python
def ADAM(x,y_2,alpha = 0.0001,eps = 0.00000001,iterations = 10000):
    beta1 = .9
    beta2 = 0.999
    cost_list = []
    
    theta_0_list = []
    theta_1_list = []
    
    grad_0_list = []
    grad_1_list = []
    
    hypothesis_list=[]
    
    #theta intilization
    theta_0 = 0
    theta_1 = 0
    
    prev_v_theta_0 = 0
    prev_v_theta_1 = 0
    
    prev_m_theta_0 = 0
    prev_m_theta_1 = 0
    
    m = len(y_2)
    for i in range (iterations):
        
        #hypothesis
        h = theta_0 + (theta_1 * x)
        hypothesis_list.append(h)
        
        #get cost
        cost = J(x, y_2, theta_0 , theta_1)
        cost_list.append(cost)
        
        # caluulate gradiant
        
#         grad_0 = (( h - y ).sum())/m
#         grad_1= ((( h - y )* x) .sum())/m
        g0 = dx_theta_0(theta_0,theta_1,x,y_2)
        g1 = dx_theta_1(theta_0,theta_1,x,y_2)
        
        grad_0_list.append(g0)
        grad_1_list.append(g1)
        
        
        m_theta_0 = beta1*prev_m_theta_0  + (1-beta1)*(g0)
        m_theta_1 = beta1*prev_m_theta_1  + (1-beta1)*(g1)
        
        v_theta_0 = beta2*prev_v_theta_0  + (1-beta2)*(g0 * g0)
        v_theta_1 = beta2*prev_v_theta_1  + (1-beta2)*(g1 * g1)
        
        prev_m_theta_0 = m_theta_0
        prev_m_theta_1 = m_theta_1
        
        prev_v_theta_0 = v_theta_0
        prev_v_theta_1 = v_theta_1
        
        #update thetas
        theta_0 = theta_0 - (alpha/(np.sqrt(prev_v_theta_0)+eps )) * prev_m_theta_0
        theta_1 = theta_1 - (alpha/(np.sqrt(prev_v_theta_1) + eps))  * prev_m_theta_1
        
        theta_0_list.append(theta_0)
        theta_1_list.append(theta_1)
        
        
        #if ther's no change in the cost so break
        if i > 1:
            if abs(cost_list[i-1]- cost_list[i])<0.0001:
                break

            #if ther's no change in gradiant so, break
            if abs(np.linalg.norm([theta_0_list[i],theta_1_list[i]])-np.linalg.norm([theta_0_list[i-1],theta_1_list[i-1]]))< 0.0001:
                break

            #if no chane in thetas so break
            if abs(np.linalg.norm([grad_0_list[i],grad_1_list[i]])-np.linalg.norm([grad_0_list[i-1],grad_1_list[i-1]])) < 0.0001:
                break
    
    return theta_0 , theta_1, cost_list, theta_0_list, theta_1_list, hypothesis_list
```
